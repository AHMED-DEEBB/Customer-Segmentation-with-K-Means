1. Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

2. Generate Sample Data (Customer Spending Patterns)

```python
# simulate customers using make_blobs, which creates blobs (natural clusters).

# create synthetic customer data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Optional: Normalize the data (not required for blobs but useful in real datasets)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

3. Apply K-Means Clustering

```python
# fit kmeans model
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X_scaled)

# get clusters labels and centriods
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

4. Visualize the Clusters

```python
# plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, alpha=0.5, marker='X', label='Centroids')
plt.title('Customer Segmentation with K-Means')
plt.xlabel('Feature 1 (e.g., Annual Spending)')
plt.ylabel('Feature 2 (e.g., Visit Frequency)')
plt.legend()
plt.grid(True)
plt.show()
```
